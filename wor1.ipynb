```python
#1
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')

plt.title("title 1")
plt.xlabel("x label",color="blue")
plt.ylabel("y label",color="blue")
plt.show()
```

```python
#2
import matplotlib.pyplot as plt
import numpy as np
y1 = np.array([2, 7, 2, 7])
y2 = np.array([7, 2, 7, 2])
plt.title("GRAPH")
plt.plot(y1,marker="*",lw="2.5")
plt.plot(y2,linestyle="dashed",lw="2.5")
plt.show()
```

```python
#4
import numpy as np
import matplotlib.pyplot as plt
x = np.array([2,4,6,8])
y = np.array([3,5,7,9])

plt.title("title 2")
plt.plot(x, y)
plt.grid()
plt.show() 
```

```python
#5
import matplotlib.pyplot as plt
import numpy as np


x = np.array([2,2,8,1,15,])
y = np.array([100,105,84,105,90])

sizes = np.array([215,146,543,362,100])
color=np.array([20,40,63,14,21])
plt.scatter(x, y, c=color,s=sizes)

plt.show() 
```

```python
#6
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,2,3,4])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red",width=0.1)
plt.show()
```

```python
#7
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["35","25","20","20"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, colors=mycolors,shadow = True)
plt.legend()
plt.show() 
```
